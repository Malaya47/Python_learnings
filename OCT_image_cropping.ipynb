```
from google.colab import files
uploaded = files.upload()
```

```
import cv2
import numpy as np
import os

os.makedirs("processed", exist_ok=True)

for idx, filename in enumerate(uploaded.keys()):
    img = cv2.imread(filename)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    h, w = img.shape
    scale = 512 / max(h, w)

    new_h = int(h * scale)
    new_w = int(w * scale)

    resized = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_AREA)

    canvas = np.zeros((512, 512), dtype=np.uint8)

    y_offset = (512 - new_h) // 2
    x_offset = (512 - new_w) // 2

    canvas[y_offset:y_offset+new_h, x_offset:x_offset+new_w] = resized

    cv2.imwrite(f"processed/oct_{idx:03d}.png", canvas)

print("Done processing.")
```

```
import shutil

shutil.make_archive("oct_processed", 'zip', "processed")
files.download("oct_processed.zip")
```
